# Import and process the data

```python
### z-score all variables except for quality

# find the columns we want to normalize (all except quality)
cols2zscore = data.keys()
cols2zscore = cols2zscore.drop('quality')

# z-score (written out for clarity)
for col in cols2zscore:
  meanval   = np.mean(data[col])
  stdev     = np.std(data[col],ddof=1)
  data[col] = (data[col]-meanval) / stdev

# can also do more compactly
#data[cols2zscore] = data[cols2zscore].apply(stats.zscore)

data.describe()
```

# Re-organize the data: train/test in DataLoaders

```python
# convert from pandas dataframe to tensor
dataT  = torch.tensor( data[cols2zscore].values ).float()
labels = torch.tensor( data['boolQuality'].values ).float()

print( dataT.shape )
print( labels.shape )

# we'll actually need the labels to be a "tensor"
labels = labels[:,None]
print( labels.shape )
```

```python
# use scikitlearn to split the data
train_data,test_data, train_labels,test_labels = train_test_split(dataT, labels, test_size=.1)


# then convert them into PyTorch Datasets (note: already converted to tensors)
train_data = TensorDataset(train_data,train_labels)
test_data  = TensorDataset(test_data,test_labels)


# finally, translate into dataloader objects
batchsize    = 64
train_loader = DataLoader(train_data,batch_size=batchsize,shuffle=True, drop_last=True)
test_loader  = DataLoader(test_data,batch_size=test_data.tensors[0].shape[0])
```

```python
# prompt: Loop over all the variables in the dataset, #    and perform an independent-samples t-test on the data for the binarized wine quality. Which variables are significantly #    different between "low" and "high" quality wine?

import scipy.stats as stats

for col in data.keys():
  if col in ["boolQuality", "quality"]:
    continue
  # perform independent-samples t-test
  tstat, pval = stats.ttest_ind(data[col][data['boolQuality']==0], data[col][data['boolQuality']==1])
  # print results
  print(f"{col}: t-statistic = {tstat:.3f}, p-value = {pval:.3f}")
  # interpret results
  if pval < 0.05:
    print(f"{col} is significantly different between low and high quality wine.")
  else:
    print(f"{col} is not significantly different between low and high quality wine.")
```

```python
# prompt: Use pytorch to create a model to predict `QualityBool` (values 0 or 1) from the columns in the datasets (excluding the `Quality` column`

import torch
from torch import nn

# Define the model
class WineQualityModel(nn.Module):
  def __init__(self):
    super(WineQualityModel, self).__init__()
    # Define the layers of the model
    self.layers = nn.Sequential(
      nn.Linear(11, 128),
      nn.ReLU(),
      nn.Linear(128, 64),
      nn.ReLU(),
      nn.Linear(64, 32),
      nn.ReLU(),
      nn.Linear(32, 16),
      nn.ReLU(),
      nn.Linear(16, 1),
      #nn.Sigmoid()
    )

  def forward(self, x):
    # Forward pass through the model
    x = self.layers(x)
    return x

def train_the_model(model, train, test, epochs):
  # Train the model
  trainAcc = []
  testAcc  = []
  losses   = []
  for epoch in range(epochs):
    batchAcc  = []
    batchLoss = []
    for inputs, labels in train:
      # Forward pass
      outputs = model(inputs)
      loss = criterion(outputs, labels)
      batchAcc.append( 100*torch.mean(((outputs>0) == labels).float()).item())
      batchLoss.append( loss.item() )

      # Backward pass
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

    # test accuracy
    model.eval()
    with torch.no_grad():
      X,y = next(iter(test)) # extract X,y from test dataloader
      yHat = model(X)
      testAcc.append( 100*torch.mean(((yHat>0) == y).float()).item() )

    # Print the loss after every 10 epochs
    if epoch % 100 == 0:
      print(f'Epoch {epoch+1}/{100}, Loss: {loss.item():.4f}')
    trainAcc.append( np.mean(batchAcc) )
    losses.append(np.mean(batchLoss))

  return trainAcc, testAcc, losses




# Instantiate the model
model = WineQualityModel()

# Define the loss function and optimizer
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# translate into dataloader objects
numepochs = 500
batchsize    = 16
train_loader = DataLoader(train_data,batch_size=batchsize,shuffle=True, drop_last=True)
test_loader  = DataLoader(test_data,batch_size=test_data.tensors[0].shape[0])

trainAcc, testAcc, losses = train_the_model(model=model, train=train_loader, test=test_loader, epochs=numepochs)
```

```python
# plot the results
fig,ax = plt.subplots(1,2,figsize=(15,5))


ax[0].plot(losses,'k^-')
ax[0].set_ylabel('Loss')
ax[0].set_xlabel('Epochs')
ax[0].set_title('Losses with minibatch size=' + str(batchsize))

ax[1].plot(trainAcc,'ro-')
ax[1].plot(testAcc,'bs-')
ax[1].set_title('Accuracy with minibatch size=' + str(batchsize))
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy (%)')
ax[1].legend(['Train','Test'])
ax[1].set_ylim([27,103])

plt.show()
```

```python
import time

batch_sizes = 2**np.arange(1,7, 1)
numepochs = 500

trainAccs = np.zeros((len(batch_sizes), numepochs))
testAccs = np.zeros((len(batch_sizes), numepochs))
mean_results = np.zeros((len(batch_sizes), 2))
times = []

for bi, batch in enumerate(batch_sizes):
  start_time = time.time()
  # set up batch size (train only)
  # finally, translate into dataloader objects
  model = WineQualityModel()
  train_loader = DataLoader(train_data,batch_size=int(batch),shuffle=True,drop_last=True)

  trainAcc,testAcc,losses = train_the_model(model=model, train=train_loader, test=test_loader, epochs=numepochs)
  # store accuracies
  mean_results[bi,0] = np.mean(trainAcc[-50:])
  mean_results[bi,1] = np.mean(testAcc[-50:])

  end_time = time.time()
  trainAccs[bi, :] = trainAcc
  testAccs[bi, :] = testAcc
  times.append(end_time - start_time)
```

```python
# plot the experiment results
fig,ax = plt.subplots(1,2,figsize=(15,5))


ax[0].plot(batch_sizes,mean_results,'o-')
ax[0].set_xlabel('Batch Size')
ax[0].set_ylabel('Average accuracy')
ax[0].legend(['Train','Test'])

ax[1].plot(batch_sizes,-np.diff(mean_results,axis=1),'o-')
ax[1].plot([0,.9],[0,0],'k--')
ax[1].set_xlabel('Batch Size')
ax[1].set_ylabel('Train-test difference (acc%)')

plt.show()
```

```python
# plot the experiment results
fig,ax = plt.subplots(1,2,figsize=(15,5))

cmaps = plt.cm.plasma(np.linspace(1, 0, len(batch_sizes)))

for bi, batch in enumerate(batch_sizes):
  ax[0].plot(trainAccs[bi], color=cmaps[bi], label=batch)
  ax[1].plot(testAccs[bi], color=cmaps[bi], label=batch)


ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Train Accuracy')
ax[0].legend()

# ax[1].plot([0,.9],[0,0],'k--')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Test Accuracy')
ax[1].legend()

plt.show()
```
